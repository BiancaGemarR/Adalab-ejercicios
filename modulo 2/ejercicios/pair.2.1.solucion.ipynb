```python
url = "https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date=2014-01-01T00:00&end_date=2014-12-31T23:59&time_trunc=month"
```

```python
diccionario_año = {"tipo": [], "titulo": [], "fecha":[], "valor":[]}

for a in range(2011, 2021):
    url = f"https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date={a}-01-01T00:00&end_date={a}-12-31T23:59&time_trunc=month"
    res = requests.get(url)
    print(url)
    datos = res.json()
    for e in datos["included"]:
    
        for dato in e["attributes"]['values']:

            diccionario_año["titulo"].append(e["attributes"]["title"])
            diccionario_año["tipo"].append(e["attributes"]["type"])
            diccionario_año["fecha"].append(dato["datetime"])
            diccionario_año["valor"].append(dato["value"])
```
